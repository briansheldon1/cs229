```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
```

```python
# Create Simulated Gaussian Data

# number of features in x
num_features = 2

# randomly create correlation matrices for x0 and x1
x0_corrs = np.random.rand(num_features, num_features)
x0_corrs = x0_corrs.T @ x0_corrs + np.eye(num_features)*1e-3

x1_corrs = np.random.rand(num_features, num_features)
x1_corrs = x1_corrs.T @ x1_corrs + np.eye(num_features)*1e-3

# define means and variances for the two labels
x0_means = np.random.randn(num_features)
x1_means = np.random.randn(num_features)

# create data
m0 = 1000
m1 = 1000
x0_data = np.random.multivariate_normal(x0_means, x0_corrs, m0)
x1_data = np.random.multivariate_normal(x1_means, x1_corrs, m1)
```

```python
# Train parameters

# phi = E[y]
phi = m1/(m0+m1)

# feature averages
mu0 = np.sum(x0_data, axis=0)/m0
mu1 = np.sum(x1_data, axis=0)/m1

# single covariance
x_all = np.vstack((x0_data-mu0, x1_data-mu1))
cov = np.dot(x_all.T, x_all)/(m0+m1)
cov_inv = np.linalg.inv(cov)

# two covariances
cov0 = np.dot((x0_data-mu0).T, (x0_data-mu0))/m0
cov0_inv = np.linalg.inv(cov0)
cov1 = np.dot((x1_data-mu1).T, (x1_data-mu1))/m1
cov1_inv = np.linalg.inv(cov1)
```

```python
# Test model

def pred_onecov(x, mu0, mu1, phi, cov_inv):
    """
    Predicts the label using single covariance matrix
    """
    # Compute log probabilities (without exponentiation)
    log_p0 = -0.5 * (x - mu0).T @ cov_inv @ (x - mu0) + np.log(1 - phi)
    log_p1 = -0.5 * (x - mu1).T @ cov_inv @ (x - mu1) + np.log(phi)

    # Return class label
    return 1 if log_p1 > log_p0 else 0

def pred_twocov(x, mu0, mu1, phi, cov0, cov0_inv, cov1, cov1_inv):
    
    # Compute log probabilities (without exponentiation)
    log_p0 = -0.5 * (x - mu0).T @ cov0_inv @ (x - mu0) \
              + np.log(1 - phi) - 0.5*np.log(np.linalg.det(cov0))
    log_p1 = -0.5 * (x - mu1).T @ cov1_inv @ (x - mu1) \
              + np.log(phi) - 0.5*np.log(np.linalg.det(cov1))

    # Return class label
    return 1 if log_p1 > log_p0 else 0
    

# get predictions for single covariance
x0_preds = np.array([pred_onecov(x, mu0, mu1, phi, cov_inv) for x in x0_data])
x1_preds = np.array([pred_onecov(x, mu0, mu1, phi, cov_inv) for x in x1_data])
errors_1cov = np.sum(x0_preds==1) + np.sum(x1_preds==0)

# get predictions for two covariance
x0_preds = np.array([pred_twocov(x, mu0, mu1, phi, cov0, cov0_inv, cov1, cov1_inv) for x in x0_data])
x1_preds = np.array([pred_twocov(x, mu0, mu1, phi, cov0, cov0_inv, cov1, cov1_inv) for x in x1_data])
errors_2cov = np.sum(x0_preds==1) + np.sum(x1_preds==0)

# Report errors
print(f"One Cov Matrix Percent Errors: {round(100*errors_1cov/(m0+m1), 3)}%")
print(f"Two Cov Matrix Percent Errors: {round(100*errors_2cov/(m0+m1), 3)}%")
```

```python
# Plot for two of the variables

# plot data
fig, ax = plt.subplots()
ax.scatter(x0_data[:,0], x0_data[:, 1], color="blue", label="y=0")
ax.scatter(x1_data[:,0], x1_data[:, 1], color="red",  label="y=1")
x_limits, y_limits = ax.get_xlim(), ax.get_ylim()
ax.set_title("Data")
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.legend()
plt.savefig("plots/Data.png")
plt.show()

# Create testing data (only works for 2 feature problem)
numx, numy = 500, 500 #(total points = numx*numy)
grid_x, grid_y = np.meshgrid(np.linspace(*x_limits, numx), np.linspace(*y_limits, numy))
points = np.column_stack([grid_x.ravel(), grid_y.ravel()])
pred_1cov = np.array([pred_onecov([x,y], mu0, mu1, phi, cov_inv) for x,y in points])
pred_2cov = np.array([pred_twocov([x,y], mu0, mu1, phi, cov0, cov0_inv, cov1, cov1_inv) for x,y in points])

# Plot predictions for one covariance matrix
pred0 = points[pred_1cov==0, :]
pred1 = points[pred_1cov==1, :]
fig, ax = plt.subplots()
ax.scatter(pred0[:,0], pred0[:,1], color="blue", label="pred0")
ax.scatter(pred1[:,0], pred1[:,1], color="red",  label="pred1")
ax.legend()
ax.set_title("One Covariance Decision Boundary")
ax.set_xlabel("x0")
ax.set_ylabel("x1")
plt.savefig("plots/1CovBoundary.png")
plt.show()

# Plot predictions for two covariance matrices
pred0 = points[pred_2cov==0]
pred1 = points[pred_2cov==1]
fig, ax = plt.subplots()
ax.scatter(pred0[:,0], pred0[:,1], color="blue", label="pred0")
ax.scatter(pred1[:,0], pred1[:,1], color="red",  label="pred1")
ax.legend()
ax.set_title("Two Covariance Decision Boundary")
ax.set_xlabel("x0")
ax.set_ylabel("x1")
plt.savefig("plots/2CovBoundary.png")
plt.show()
```

```python
# Plot P(y=1|x), P(y=0|x)
import timeit

# Get Probabilities
def get_probs_vec(points, mu0, mu1, phi, cov_inv):
    # x is now shape (num_samples, num_features)
    diff0 = points - mu0   # shape: (num_samples, num_features)
    diff1 = points - mu1

    log_p0 = -0.5 * np.sum(diff0 @ cov_inv * diff0, axis=1) + np.log(1 - phi)
    log_p1 = -0.5 * np.sum(diff1 @ cov_inv * diff1, axis=1) + np.log(phi)

    p0, p1 = np.exp(log_p0), np.exp(log_p1)
    c = p0 + p1
    p0 = p0/c
    p1 = p1/c
    return np.column_stack((p0, p1))  # shape: (num_samples, 2)

probs = get_probs_vec(points, mu0, mu1, phi, cov_inv)

# Plot p(y=1|x)
fig, ax = plt.subplots(2, 1, figsize=(6, 10))
ax[0].plot(points[:, 0], probs[:, 1])
ax[0].set_title("Prob(y=1 | [x0, x1=0] for Single Covariance")
ax[0].set_xlabel("x0")
ax[0].set_ylabel("Prob(y=1)")

# Plot p(y=0|x)
ax[1].plot(points[:, 0], probs[:, 0])
ax[1].set_title("Prob(y=0 | [x0, x1=0] for Single Covariance")
ax[1].set_xlabel("x0")
ax[1].set_ylabel("Prob(y=1)")
plt.savefig("plots/y_probs.png")
plt.show()
```
